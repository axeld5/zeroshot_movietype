```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from openai import OpenAI
client = OpenAI()

def get_embedding(text, model="text-embedding-3-small"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
import pandas as pd

movie_df = pd.read_csv("data/imdb_dataset.csv")
sampled_data = movie_df["review"].sample(50).values.tolist()
sampled_embeddings = [get_embedding(data) for data in sampled_data]
```

```python
movie_types = pd.read_excel("movietype.xlsx")
type_list = movie_types["Movie_Type"].values.tolist()
desc_list = movie_types["Description"].values.tolist()
desc_embeddings = [get_embedding(data) for data in desc_list]
```

```python
import numpy as np

for i, (text, emb) in enumerate(zip(sampled_data, sampled_embeddings)):
    diff_array = np.array([np.linalg.norm(np.array(desc_embedding) - np.array(emb)) for desc_embedding in desc_embeddings])
    chosen_type = type_list[np.argmin(diff_array)]
    print(text)
    print(chosen_type)
```
